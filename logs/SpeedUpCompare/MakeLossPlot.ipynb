```python
import numpy as np 
import matplotlib.pyplot as pp
import math
```

```python
def moving_average(a, n) :
    if n<25 :
        Window=int(math.ceil(0.5*n))
    else :
        Window=25
    return np.sum(a[-Window-1:-1])/Window
```

```python
SpeedUpData20=np.loadtxt("SpeedUpData20.txt")
n_steps20=len(SpeedUpData20)
count20=np.arange(0,n_steps20)
AVGSpeedUp20=np.zeros((n_steps20,1))
for j in range(0,n_steps20):
    AVGSpeedUp20[j]=moving_average(SpeedUpData20[:j],j)

SpeedUpData25=np.loadtxt("SpeedUpData25.txt")
n_steps25=len(SpeedUpData25)
count25=np.arange(0,n_steps25)
AVGSpeedUp25=np.zeros((n_steps25,1))
for j in range(0,n_steps25):
    AVGSpeedUp25[j]=moving_average(SpeedUpData25[:j],j)
    
SpeedUpData30=np.loadtxt("SpeedUpData30.txt")
n_steps30=len(SpeedUpData30)
count30=np.arange(0,n_steps30)
AVGSpeedUp30=np.zeros((n_steps30,1))
for j in range(0,n_steps30):
    AVGSpeedUp30[j]=moving_average(SpeedUpData30[:j],j)
    
SpeedUpData35=np.loadtxt("SpeedUpData35.txt")
n_steps35=len(SpeedUpData35)
count35=np.arange(0,n_steps35)
AVGSpeedUp35=np.zeros((n_steps35,1))
for j in range(0,n_steps35):
    AVGSpeedUp35[j]=moving_average(SpeedUpData35[:j],j)

SpeedUpData40=np.loadtxt("SpeedUpData40.txt")
n_steps40=len(SpeedUpData40)
count40=np.arange(0,n_steps40)
AVGSpeedUp40=np.zeros((n_steps40,1))
for j in range(0,n_steps40):
    AVGSpeedUp40[j]=moving_average(SpeedUpData40[:j],j)    
    
    
pp.plot(count20[25:-1],AVGSpeedUp20[25:-1], "r",label="20x20 grid")
pp.plot(count25[25:-1],AVGSpeedUp25[25:-1], "b",label="25x25 grid")
pp.plot(count30[25:-1],AVGSpeedUp30[25:-1], "k",label="30x30 grid")
pp.plot(count35[25:-1],AVGSpeedUp35[25:-1], "g",label="35x35 grid")
pp.plot(count40[25:-1],AVGSpeedUp40[25:-1], "c",label="40x40 grid")

pp.xlabel('i')
pp.ylabel('GMRES/MLGMRES')
pp.title("NN Speed Up ")
pp.legend(loc='best')
pp.savefig('SpeedUp.svg')
```
