# 2D benchmarks for point source problem

These results correspond to a set up identical to that descrived in the "summary.ipynb" file. Except, now the RHS is constructed using a normalized 2D circular(two variables are uncorrelated) gaussian with $\sigma_x=\sigma_y=0.02$ . The domain is the square $[-1,1] \times [-1,1]$. The gaussian is moved along a polar rose curve with 12 petals in a non-sequential way. 

The tolerance parameters used here are as follows:

$e_1=10^{-3}$

$e_2=10^{-6}$

We loop over 2000 different linear problems for each given dimension $n$.

### Summary 

Both networks performed decently and gave roughly a 2x speed up for only a small amount of training in real-time. Generally speaking, it seems that both architectures work well the less localized the RHS is. While the 1D convolutions seem to do even worse for less localized RHS, it does considerably better  for less localized RHS.
