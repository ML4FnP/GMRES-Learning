## 1D benchmarks for linear gradient RHS problem

For this problem, the RHS is a ramp function $R(x)=max(0,x)$ whose center is constrained to move from -4 to 4 on the domain [-5,5]. Furthermore, the slope of the ramp function varies from -1 to 1. The explicit formula for the RHS vector $b$ is:

$$b_i=\left(\sin(i)+.0001 \right)\max\left(x-4\sin\left(\frac{i}{2}|\cos(\frac{i}{2})|\right),0\right)$$
where $x$ is a discretization of $[-5,5]$.
